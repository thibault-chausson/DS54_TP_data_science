__Exercice 1__

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


fibonacci(10)
```

__Exercice 2__

```python
def reverse_array(arr):
    if len(arr) == 0:
        return []
    else:
        return [arr[-1]] + reverse_array(arr[:-1])


reverse_array([1, 2, 3, 4, 5])
```

```python
def replace_odd(arr):
    if len(arr) == 0:
        return []
    else:
        if arr[0] % 2 == 0:
            return [arr[0]] + replace_odd(arr[1:])
        else:
            return [-1] + replace_odd(arr[1:])


def random_array(n):
    return [rand.randint(0, 100) for i in range(n)]


replace_odd(random_array(10))
```

__Exercice 4 : MatplotLib__

```python
#Generate a sequence of numbers from -10 to 10 with 100 steps in between
x = np.linspace(-10, 10, 100)

#Create a second array using sine
y = np.sin(x)

#Create a third array using cosine
z = np.cos(x)

#The plot function makes a line chart of one array against another
fig, axs = plt.subplots(2)
fig.suptitle('Cos and Sin')
axs[0].plot(x, y, label='sin(x)', linestyle=':')
axs[1].plot(x, z, marker="o", label='cos(x)', color='red', linewidth=5)

plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

En 3D

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline)
```

```python
# Create a Pandas dataframe from some data.
data = pd.DataFrame({'fruit': ['apple', 'orange', 'pear', 'banana'],
                     'price': [2, 1, 4, 3],
                     'stock': [10, 5, 3, 9],
                     'like': [True, False, True, False]})
data_pandas = pd.DataFrame(data)
display(data_pandas)
display(data_pandas[data_pandas.price > 2])
display(data_pandas[data_pandas.like == True])
display(data_pandas[data_pandas.stock % 3 == 0])
##the length of the fruit name is greater than 5
display(data_pandas[data_pandas.fruit.str.len() > 5])
```

__Exercice 6 : Breast Cancer__

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

cancer = load_breast_cancer()

X = cancer.data
Y = cancer.target

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0)

for i in range(1, 10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, Y_train)
    print("Accuracy of KNN n={0} on training set: {1:.2f}".format(i, knn.score(X_train, Y_train)))
```
